#### **Система Лоренца:**

Система Лоренца описується трьома диференціальними рівняннями:

$$
\begin{cases}
\frac{dx}{dt} = \sigma (y - x) \\
\frac{dy}{dt} = x (\rho - z) - y \\
\frac{dz}{dt} = xy - \beta z
\end{cases}
$$

де:
- $( \sigma )$ — коефіцієнт Прандтля (зазвичай $ \sigma = 10 $),
- $( \rho )$ — числовий параметр для турбулентності (зазвичай $ \rho = 28 $),
- $( \beta )$ — коефіцієнт відношення висоти до довжини (зазвичай $ \beta = 8/3 $).

#### **Об'єкти Нарендра-Пархтізаратхі:**

Об'єкти Нарендра-Пархтізаратхі описуються наступними рівняннями:

$$
\begin{cases}
x_1(t+1) = \alpha_1 x_1(t) + \beta_1 x_2(t) + \gamma_1 x_3(t) + \delta_1 \\
x_2(t+1) = \alpha_2 x_1(t) + \beta_2 x_2(t) + \gamma_2 x_3(t) + \delta_2 \\
x_3(t+1) = \alpha_3 x_1(t) + \beta_3 x_2(t) + \gamma_3 x_3(t) + \delta_3
\end{cases}
$$

де $(\alpha_i)$, $(\beta_i)$, $(\gamma_i)$ і $(\delta_i)$ — коефіцієнти для кожного з рівнянь.

### **Нелінійних рівнянь Нарендра-Партхізаратхі:**

### Параметри:
- $( a_1, a_2 )$, $( b_1, b_2 )$ — це коефіцієнти, які можуть бути варійовані для дослідження динаміки системи та впливу на прогнозування.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
import plotly.express as px
from scipy.integrate import solve_ivp
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
```

```python
def lorenz(xyz, parameter):
    o, r, b = parameter
    x, y, z = xyz
    x_dot = o*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return np.array([x_dot, y_dot, z_dot])


dt = 0.01
num_steps = 15000

xyzs = np.empty((num_steps + 1, 3))
xyzs[0] = (0.0, 1.0, 1.05)


parameter_sets = [
    (10, 28, 8/3),
]

dataL = []

for parameter in parameter_sets:
    for i in range(num_steps):
        xyzs[i + 1] = xyzs[i] + lorenz(xyzs[i], parameter) * dt
        
    x_vals = xyzs[:, 0]
    y_vals = xyzs[:, 1]
    z_vals = xyzs[:, 2]

    fig = px.line_3d(x=x_vals, y=y_vals, z=z_vals)
    fig.show()

    dataL.extend(zip(x_vals, y_vals, z_vals))


df = pd.DataFrame(dataL, columns=['X', 'Y', 'Z'])
csv_file_path = 'lorenz_data.csv'
df.to_csv(csv_file_path, index=False)
```

```python
a1, b1 = 1.0, 0.02
a2, b2 = 0.5, 0.05

def narendra_parthasarathy(t, y):
    x1, x2 = y
    dx1_dt = a1 * x1 * x2 - b1 * x1**3
    dx2_dt = a2 * np.sin(x1) - b2 * x2**2
    return [dx1_dt, dx2_dt]

y0 = [0.5, -0.5]

t_span = [0, 20]
t_eval = np.linspace(t_span[0], t_span[1], 13000)

sol = solve_ivp(narendra_parthasarathy, t_span, y0, t_eval=t_eval, method='RK45')

x1 = sol.y[0]
x2 = sol.y[1]
t = sol.t

df = pd.DataFrame({'t': t, 'x1': x1, 'x2': x2})
csv_file_path = 'narendra_parthasarathy_data.csv'
df.to_csv(csv_file_path, index=False)
```

```python
lorenz_data = pd.read_csv('lorenz_data.csv')



train_data, temp = train_test_split(lorenz_data, test_size=0.3, random_state=42)
val_data, test_data = train_test_split(temp, test_size=0.5, random_state=42)

print(train_data.describe())


scaler = StandardScaler()
train_data = scaler.fit_transform(train_data)
val_data = scaler.transform(val_data)
test_data = scaler.transform(test_data)


train_data = pd.DataFrame(train_data, columns=lorenz_data.columns)
val_data = pd.DataFrame(val_data, columns=lorenz_data.columns)
test_data = pd.DataFrame(test_data, columns=lorenz_data.columns)

```

```python
plt.figure(figsize=(12, 6))
plt.scatter(test_data['X'], test_data['Y'], label='Lorenz System')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, (p1, p2, p3, p4) = plt.subplots(1, 4, figsize=(15, 5))

lin_reg = LinearRegression()
dec_tree = DecisionTreeRegressor(random_state=42)
rand_forest = RandomForestRegressor(random_state=42)
svm_model = SVR()

X_train = train_data[['X', 'Y']]
y_train = train_data['Z']

lin_reg.fit(X_train, y_train)
dec_tree.fit(X_train, y_train)
rand_forest.fit(X_train, y_train)
svm_model.fit(X_train, y_train)

lin_reg_pred = lin_reg.predict(val_data[['X', 'Y']])
dec_tree_pred = dec_tree.predict(val_data[['X', 'Y']])
rand_forest_pred = rand_forest.predict(val_data[['X', 'Y']])
svm_model_pred = svm_model.predict(val_data[['X', 'Y']])


def plot_predictions(ax, x, y_actual, y_pred, title):
    ax.scatter(x, y_actual, label='Actual')
    ax.scatter(x, y_pred, label=title, alpha=0.5)
    ax.legend()
    ax.grid(True)
    ax.set_title(title)


plot_predictions(p1, val_data['X'], val_data['Z'], lin_reg_pred, 'Linear Regression')
plot_predictions(p2, val_data['X'], val_data['Z'], dec_tree_pred, 'Decision Tree')
plot_predictions(p3, val_data['X'], val_data['Z'], rand_forest_pred, 'Random Forest')
plot_predictions(p4, val_data['X'], val_data['Z'], svm_model_pred, 'SVM')

plt.show()

print("Linear Regression R²:", lin_reg.score(val_data[['X', 'Y']], val_data['Z']))
print("Decision Tree R²:", dec_tree.score(val_data[['X', 'Y']], val_data['Z']))
print("Random Forest R²:", rand_forest.score(val_data[['X', 'Y']], val_data['Z']))
print("SVM R²:", svm_model.score(val_data[['X', 'Y']], val_data['Z']))

print("----------------------------------------------------------------------------------------------------")



param_grids = {
    'dec_tree': {
        'max_depth': [3, 5, 10, None],
        'min_samples_split': [2, 5, 10]
    },
    'rand_forest': {
        'n_estimators': [10, 50, 100],
        'max_depth': [3, 5, 10, None],
        'min_samples_split': [2, 5, 10]
    },
    'svm': {
        'C': [0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2],
        'kernel': ['linear', 'rbf', 'poly']
    }
}

grid_searches = {
    'dec_tree': GridSearchCV(dec_tree, param_grids['dec_tree'], cv=5),
    'rand_forest': GridSearchCV(rand_forest, param_grids['rand_forest'], cv=5),
    'svm': GridSearchCV(svm_model, param_grids['svm'], cv=5)
}

grid_searches['dec_tree'].fit(X_train, y_train)
grid_searches['rand_forest'].fit(X_train, y_train)
grid_searches['svm'].fit(X_train, y_train)

best_dec_tree = grid_searches['dec_tree'].best_estimator_
best_rand_forest = grid_searches['rand_forest'].best_estimator_
best_svm = grid_searches['svm'].best_estimator_

lin_reg_pred = lin_reg.predict(val_data[['X', 'Y']])
dec_tree_pred = best_dec_tree.predict(val_data[['X', 'Y']])
rand_forest_pred = best_rand_forest.predict(val_data[['X', 'Y']])
svm_pred = best_svm.predict(val_data[['X', 'Y']])

fig, (p1, p2, p3, p4) = plt.subplots(1, 4, figsize=(15, 5))


plot_predictions(p1, val_data['X'], val_data['Z'], lin_reg_pred, 'Linear Regression')
plot_predictions(p2, val_data['X'], val_data['Z'], dec_tree_pred, 'Decision Tree')
plot_predictions(p3, val_data['X'], val_data['Z'], rand_forest_pred, 'Random Forest')
plot_predictions(p4, val_data['X'], val_data['Z'], svm_pred, 'SVM')




print("Linear Regression R²:", lin_reg.score(val_data[['X', 'Y']], val_data['Z']))
print("Decision Tree R² (best params):", best_dec_tree.score(val_data[['X', 'Y']], val_data['Z']))
print("Random Forest R² (best params):", best_rand_forest.score(val_data[['X', 'Y']], val_data['Z']))
print("SVM R² (best params):", best_svm.score(val_data[['X', 'Y']], val_data['Z']))
plt.show()

print("Best Decision Tree parameters found by Grid Search:", grid_searches['dec_tree'].best_params_)
print("Best Random Forest parameters found by Grid Search:", grid_searches['rand_forest'].best_params_)
print("Best SVM parameters found by Grid Search:", grid_searches['svm'].best_params_)
```

```python
print("Linear Regression R²:", r2_score(val_data['Z'], lin_reg_pred))
print("Linear Regression MSE:", mean_squared_error(val_data['Z'], lin_reg_pred))
print("Decision Tree R²:", r2_score(val_data['Z'], dec_tree_pred))
print("Decision Tree MSE:", mean_squared_error(val_data['Z'], dec_tree_pred))
print("Random Forest R²:", r2_score(val_data['Z'], rand_forest_pred))
print("Random Forest MSE:", mean_squared_error(val_data['Z'], rand_forest_pred))
print("SVM R²:", r2_score(val_data['Z'], svm_pred))
print("SVM MSE:", mean_squared_error(val_data['Z'], svm_pred))
```

```python

X_train = train_data[['t']]
y_train_x1 = train_data[['x1', 'x2']]
X_val = val_data[['t']]
y_val_x1 = val_data[['x1', 'x2']]
X_test = test_data[['t']]
y_test = test_data[['x1', 'x2']] 


xgb_model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5, subsample=0.8, colsample_bytree=0.8)
dec_tree = DecisionTreeRegressor(max_depth=10, random_state=42)
rand_forest = RandomForestRegressor(n_estimators=200, max_depth=10, random_state=42)
knn_model = LinearRegression()

xgb_model.fit(X_train, y_train_x1)
dec_tree.fit(X_train, y_train_x1)
rand_forest.fit(X_train, y_train_x1)
knn_model.fit(X_train, y_train_x1)



param_grids = {
    'xgb_model': {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [3, 5, 7],
        'subsample': [0.7, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'max_depth': [3, 4, 5]
    },
    'dec_tree': {
        'max_depth': [3, 5, 10, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    },
    'rand_forest': {
        'max_depth': [5, 10, 15],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }
}

grid_searches = {
    'xgb_model': GridSearchCV(xgb_model, param_grids['xgb_model'], cv=5, n_jobs=-1, scoring='neg_mean_squared_error'),
    'dec_tree': GridSearchCV(dec_tree, param_grids['dec_tree'], cv=5, n_jobs=-1, scoring='neg_mean_squared_error'),
    'rand_forest': GridSearchCV(rand_forest, param_grids['rand_forest'], cv=5, n_jobs=-1, scoring='neg_mean_squared_error')
}

grid_searches['xgb_model'].fit(X_train, y_train_x1)
grid_searches['dec_tree'].fit(X_train, y_train_x1)
grid_searches['rand_forest'].fit(X_train, y_train_x1)

best_xgb_model = grid_searches['xgb_model'].best_estimator_
best_dec_tree = grid_searches['dec_tree'].best_estimator_
best_rand_forest = grid_searches['rand_forest'].best_estimator_

xgb_pred = best_xgb_model.predict(X_val)
dec_tree_pred = best_dec_tree.predict(X_val)
rand_forest_pred = best_rand_forest.predict(X_val)
knn_model_pred = knn_model.predict(X_val)



def plot_predictions(ax, x, y_actual, y_pred, title):

    ax.scatter(x, y_actual[:, 0], label='Actual x1', color='blue')
    ax.scatter(x, y_pred[:, 0], label='Predicted x1', color='red')
    ax.scatter(x, y_actual[:, 1], label='Actual x2', color='green')
    ax.scatter(x, y_pred[:, 1], label='Predicted x2', color='orange')
    ax.legend()
    ax.grid(True)
    ax.set_title(title)

fig, (p1, p2, p3, p4) = plt.subplots(1, 4, figsize=(20, 5))

plot_predictions(p1, X_val['t'], y_val_x1.values, xgb_pred, 'XGBoost')
plot_predictions(p2, X_val['t'], y_val_x1.values, dec_tree_pred, 'Decision Tree)')
plot_predictions(p3, X_val['t'], y_val_x1.values, rand_forest_pred, 'Random Forest')
plot_predictions(p4, X_val['t'], y_val_x1.values, knn_model_pred, 'Linear Regression')
plt.tight_layout()
plt.show()


print("XGBoost R² test data:", r2_score(y_val_x1, xgb_pred))
print("XGBoost MSE test data:", mean_squared_error(y_val_x1, xgb_pred))
print("Decision Tree R² test data:", r2_score(y_val_x1, dec_tree_pred))
print("Decision Tree MSE test data:", mean_squared_error(y_val_x1, dec_tree_pred))
print("Random Forest R² test data:", r2_score(y_val_x1, rand_forest_pred))
print("Random Forest MSE test data:", mean_squared_error(y_val_x1, rand_forest_pred))
print("Linear Regression R² test data:", r2_score(y_val_x1, knn_model_pred))
print("Linear Regression MSE test data:", mean_squared_error(y_val_x1, knn_model_pred))

```
